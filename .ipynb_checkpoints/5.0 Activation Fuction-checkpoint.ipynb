```python
# Import resources
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Linear
x = np.linspace(-5, 5, 100) # Prepare the data
y = x*0 # function

# Plot the data
plt.plot(x, y, label='Linear')
plt.legend() # Add a legend
plt.show()# Show the plot
```

```python
# Sigmoid
x = np.linspace(-5, 5, 100) # Prepare the data
y = 1/(1 + np.exp(-x)) # function

# Plot the data
plt.plot(x, y, label='Sigmoid')
plt.legend() # Add a legend
plt.show()# Show the plot
```

```python
# Hyperbolic Tangent
x = np.linspace(-5, 5, 100) # Prepare the data
y = np.tanh(x) # function

# Plot the data
plt.plot(x, y, label='Hyperbolic Tangent')
plt.legend() # Add a legend
plt.show()# Show the plot
```

```python
    'Null': lambda x: x*0,
    'Sigmoid': lambda x: 1/(1 + np.exp(-x)),
    'Hyperbolic Tangent': lambda x: np.tanh(x),
    'Cosine': lambda x: np.cos(x),
    'Gaussian': lambda x: np.exp(-x**2/2),
```

```python
# Cosine
x = np.linspace(-5, 5, 100) # Prepare the data
y = np.cos(x) # function

# Plot the data
plt.plot(x, y, label='Cosine')
plt.legend() # Add a legend
plt.show()# Show the plot
```

```python
# Gaussian
x = np.linspace(-5, 5, 100) # Prepare the data
y = np.exp(-x**2/2) # function

# Plot the data
plt.plot(x, y, label='Gaussian')
plt.legend() # Add a legend
plt.show()# Show the plot
```
